```python
def f(r):
    import math
    a = 1
    b = .001
    return a * math.exp(-b * r) / r**3
```

```python
f(.5)
```

```python
f(5)
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
X = np.arange(.1, 5, .001)
Y = [f(x) for x in X]

plt.xlabel("Force (Verruckts)")
plt.ylabel("Attractive Force (Fermis)")
plt.plot(X, Y)
```

```python
### We'll integrate under the graph from .5 - 5
from scipy.integrate import quad
quad(f, .5, 5)
```
